```python
import pandas as pd
import matplotlib.pyplot as plt

# File paths for your CSVs  replace these with your actual model paths
csv_files = {
    "EfficientNetV2-S": "/90daydata/nematode_ml/BLD/nematode_project/outputs/5 epoch wait/efficientnet_metrics_2025-09-08_13-04-28.csv",
    "InceptionV3": "/90daydata/nematode_ml/BLD/nematode_project/outputs/5 epoch wait/inception_metrics_2025-09-08_14-08-25.csv",
    "MobileNetV3-L": "/90daydata/nematode_ml/BLD/nematode_project/outputs/5 epoch wait/mobilenet_metrics_2025-12-12_10-26-12.csv",
    "ResNet50": "/90daydata/nematode_ml/BLD/nematode_project/outputs/5 epoch wait/resnet_metrics_2025-09-08_14-51-25.csv"
}

# Set global font sizes
plt.rcParams.update({
    'axes.titlesize': 20,   # subplot titles
    'axes.labelsize': 18,   # x/y labels
    'xtick.labelsize': 16,  # x tick labels
    'ytick.labelsize': 16,  # y tick labels
    'legend.fontsize': 14,  # legend text
})

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

from matplotlib.ticker import MaxNLocator

# Decide how many ticks you want (e.g., 6 tick marks including start and end)
num_ticks = 6  

for ax, (model_name, file_path) in zip(axes, csv_files.items()):
    df = pd.read_csv(file_path)
    
    ax.plot(df['epoch'], df['train_acc'] * 100, label='Training Accuracy', linewidth=3)
    ax.plot(df['epoch'], df['val_acc'] * 100, label='Validation Accuracy', linewidth=3)
    
    ax.set_title(f'{model_name}')
    ax.set_xlabel('Epoch')
    ax.set_ylabel('Accuracy (%)')
    ax.set_ylim(60, 100)
    ax.grid(True)
    ax.legend(loc='lower right')
    
    #  Set roughly equal number of ticks, but adapt to the epoch range
    ax.xaxis.set_major_locator(MaxNLocator(nbins=num_ticks, integer=True))



# At the very end of your plotting code, before plt.show()
plt.tight_layout()

# Save the figure as a PNG file (300 dpi for crisp text/lines)
plt.savefig("model_accuracy_comparison.png", dpi=300)

plt.show()
```

```python
# 2. Plot Loss
plt.figure(figsize=(10, 5))
plt.plot(df['epoch'], df['train_loss'], label='Training Loss')
plt.plot(df['epoch'], df['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training vs Validation Loss')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 3. Plot Accuracy
plt.figure(figsize=(10, 5))
plt.plot(df['epoch'], df['train_acc']* 100, label='Training Accuracy')
plt.plot(df['epoch'], df['val_acc'] * 100, label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.title('Training vs Validation Accuracy')
plt.legend()
plt.ylim(0, 105)
plt.grid(True)
plt.show()
```
